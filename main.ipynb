```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_pickle("Dataframe3.pkl")

acc = df["Acc"].tolist()[0]
new = []

for index,item in enumerate(acc):
    new.append(acc[index][0])

def plot_line(x_size, y,figname):
    x = np.arange(0,x_size,1)
    plt.figure(figsize=(12,4))
    plt.ylim([0.7,1.3])
    fig = plt.plot(x,y,linewidth = 1)

    plt.savefig("{}.png".format(figname))
    plt.show()


plot_line(4760,new,"inteiro")
```

```python

def get_interval_from_activity_id(activity_id : int ) -> float:
    if activity_id < 23:
        return 0
    if activity_id > 100:
        return 7.5
    return 5

def time_to_sample(time: int) -> int:
    return int(time * 238) 

def trimVectorValues(df):
    for index, row in df.iterrows():
        acc = row["Acc"]
        activity_id = row["ActivityID"]

        start = get_interval_from_activity_id(activity_id=activity_id)
        end = start + 5

        start_sample = time_to_sample(start)
        end_sample = time_to_sample(end)

        df.at[index, "Acc"] = acc[start_sample:end_sample]    

trimVectorValues(df)

acc = df["Acc"].tolist()[0]
new = [] 

zeros = np.zeros(4760-1190).tolist()


for index,item in enumerate(acc):
    new.append(acc[index][0])

new2 = new + zeros

plot_line(4760,new2,"cortado")

df.to_csv("test1.csv")

from statistics import mean
b = [ 3, 5 , 7 , 10]

print(mean(b))
```
